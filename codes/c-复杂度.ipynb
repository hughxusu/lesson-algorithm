```python
%%time
for a in range(1001):
    for b in range(1001):
        c = 1000 - a - b
        if a**2 + b**2 == c**2:
            print(a, b, c)
```

```python
import time

for i in range(9):
    n = pow(10, i)
    start = time.time()
    sum = 0
    for j in range(n):
        sum += j
    end = time.time()

    print(f'{n}: {end - start}')
```

```python
for i in range(5):
    n = pow(10, i)
    start = time.time()
    sum = 0
    for j in range(n):
        for k in range(n):
            sum += k
    end = time.time()

    print(f'{n}: {end - start}')
```
